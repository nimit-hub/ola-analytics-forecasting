exporting more than one table for power bi dashboard making 

```python
import sqlite3
import pandas as pd

db_path = r"C:\Users\Nimit\OneDrive\Desktop\Ola Project\db\ola_analytics.db"
output_dir = r"C:\Users\Nimit\OneDrive\Desktop\Ola Project\data"

conn = sqlite3.connect(db_path)

tables = ["rides", "customers", "locations"]

for table in tables:
    df = pd.read_sql(f"SELECT * FROM {table};", conn)
    df.to_csv(f"{output_dir}/{table}.csv", index=False)
    print(f"{table}.csv exported")

conn.close()
```
